```python
from ultralytics import YOLO # here importing yolo file for neural network of cnn(convolutional neural network)
import pyttsx3
from gtts import gTTS
import os
from IPython.display import Audio
```

```python
model = YOLO("yolov8m.pt")
```

```python
results = model.predict("./Dog.png")
```

# New Section

```python
result = results[0]
```

```python
len(result.boxes)
```

```python
box = result.boxes[0]
```

```python
print("Object type:", box.cls)

print("Coordinates:", box.xyxy)
print("Probability:", box.conf)
```

```python
cords = box.xyxy[0].tolist()
class_id = box.cls[0].item()
conf = box.conf[0].item()
print("Object type:", class_id)
print("Coordinates:", cords)
print("Probability:", conf)
```

```python
for x in result.names.items():
  if(x[1] == "train"):
    print(True)

print(result.names)
```

```python
cords = box.xyxy[0].tolist()
cords = [round(x) for x in cords]
class_id = result.names[box.cls[0].item()]
conf = round(box.conf[0].item(), 2)
print("Object type:", class_id)
print("Coordinates:", cords)
print("Probability:", conf)
```

```python
text = f"{class_id} detected with {int(conf * 100)} percent confidence"
tts = gTTS(text=text, lang='en')
tts.save("output.mp3")

# Play the audio in the notebook
Audio("output.mp3", autoplay=True)
```
